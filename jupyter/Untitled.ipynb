```python
import os
size_path = os.path.join(r"C:\Users\xinfahux\Downloads", "size_ipex.txt")
def get_size(gitlog):
    if os.path.isfile(gitlog):
        contents = []
        with open(gitlog, "r") as f:
            for line in f.readlines():
                line = line.strip().split("\t")
                contents.append([os.path.basename(line[-1]).strip(), line[0].strip()])
        return contents
    return

def upload_infux_db(payload):
    client = InfluxDBClient(host='mlp-trigger-1.sh.intel.com', port=8086)
    client.switch_database("mydb")
    client.write_points(payload)
    
size_list = get_size(size_path)
payload = []
if size_list:
    for index,line in enumerate(size_list):
        if ".whl" in line[0]:
            whl_name = line[0]
            whl_size = float(line[1].split("M")[0])
            lib_size = float(size_list[index+1][1].split("M")[0])
            payload.append(
                {
                    "measurement": "size_ipex5",
                    "tags": {"wheel_name": whl_name},
                    "fields": {"wheel_size": whl_size, 'libipex_gpu_core.so': lib_size,
                               "branch": "master", "commit": "d9490f048dc3b77a453lpo45"}
                }
            )
print(payload)
upload_infux_db(payload)
```

```python
table_template = '''
<div class="body">
<h2>{{title}}</h2>
<table width="98%" style="text-align:center;border:2px solid black;border-collapse:collapse;">
    <thead>
        <tr>
            {% for head in heads %}
            <th style="background-color:#0171c5;text-align:center;border:2px solid black;">{{head}}</th>
            {% endfor %}
        </tr>
    </thead>
    <tbody>
        {% for row in contents %}
            <tr>
                {% for cell in row %}
                    {% if cell and "FAIL" in (cell | string) %}
                        <td style="text-align:center;border:2px solid black;background-color:#FF9797;">{{cell}}</td>
                    {% elif cell and "PASS" in (cell | string) %}
                        <td style="text-align:center;border:2px solid black;background-color:#A6FFA6;">{{cell}}</td>
                    {% else %}
                        <td style="text-align:center;border:2px solid black;">{{cell}}</td>
                    {% endif %}
                {% endfor %}
            </tr>
        {% endfor %}
    </tbody>
</table>
</div>
'''
```

```python
title = 'hehe'
heads = ['name', 'age']
contents = [['xinfa', 20], ['tom', 23], ['PASS',90],['FAIL',78]]
html_string = jinja2.Template(table_template).render(title=title, heads=heads, contents=contents)
html_string
```
